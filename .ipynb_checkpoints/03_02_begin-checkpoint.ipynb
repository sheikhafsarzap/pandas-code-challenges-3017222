# Parsing datetime data

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/landslides.csv")
df.head()
```

## Check and handle missing values

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df[~df['date'].isnull()]
```

```python
df
```

```python
df['time'].value_counts()
```

```python
df['time'].isna().sum()
```

```python
df['time'] = df['time'].fillna("Not Known")
df
```

```python
df['time'].isna().sum()
```

```python
mean = df['fatalities'].mean()
mean
```

```python
df['fatalities'] = df['fatalities'].fillna(mean)
df
```

```python
df['fatalities'].isna().sum()
```

## Date time parsing using .to_datetime()

```python
df['date']
```

```python
df['parsed_date']= pd.to_datetime(df['date'],format="%m/%d/%y")
```

```python
df
```

```python
month_of_is_slide = df['parsed_date'].dt.month
month_of_is_slide
```

```python
import seaborn as sns
```

```python
month_of_is_slide = month_of_is_slide.dropna()
sns.histplot(month_of_is_slide, kde=False, bins=12)
```
